```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
hr_data = pd.read_csv('HR Data.csv')
print(hr_data.head())
print(hr_data.info())
```

```python
hr_data.columns
```

```python
#Data cleaning
unnecessary_columns = ['EmployeeCount', 'EmployeeNumber', 'StandardHours']
for column in unnecessary_columns:
    if column not in hr_data.columns:
        print(f"Column '{column}' not found in hr_data.")
```

```python
# Remove unnecessary columns
existing_columns = [column for column in unnecessary_columns if column in hr_data.columns]
hr_data = hr_data.drop(columns=existing_columns, axis=1)
hr_data = hr_data.drop(columns=unnecessary_columns, axis=1, errors='ignore')
print(hr_data.drop)
```

```python
#rename the columns
new_column_names = {
    'DistanceFromHome': 'DistanceFromHome',
    'Education': 'Education',
    'EducationField': 'EducationField',
    'RelationshipSatisfaction': 'RelationshipSatisfaction',
    'StockOptionLevel': 'StockOptionLevel',
    'TotalWorkingYears': 'TotalWorkingYears',
    'TrainingTimesLastYear': 'TrainingTimesLastYear',
    'WorkLifeBalance': 'WorkLifeBalance',
    'YearsAtCompany': 'YearsAtCompany',
    'YearsInCurrentRole': 'YearsInCurrentRole',
    'YearsSinceLastPromotion': 'YearsSinceLastPromotion',
    'YearsWithCurrManager': 'YearsWithCurrManager',
    'Attrition': 'Attrition',
    'BusinessTravel': 'BusinessTravel',
    'Department': 'Department',
    'Age': 'Age'
}
hr_data = hr_data.rename(columns=new_column_names)
print(new_column_names)
```

```python
# Eliminate redundant entries (if any)
duplicate_val= hr_data.drop_duplicates()
```

```python
# Sanitize specific columns (if necessary)
# Example: Convert categorical columns to appropriate
hr_data['Attrition'] = hr_data['Attrition'].map({'Yes': 1, 'No': 0})
print(hr_data['Attrition'])
```

```python
# Visualize HR data analytics for each attribute
for column in hr_data.columns:
    if hr_data[column].dtype in ['int64', 'float64']:
        plt.figure(figsize=(8, 6))
        sns.histplot(hr_data[column], kde=True, color='skyblue')
        plt.title(f'{column} Distribution')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
    elif hr_data[column].dtype == 'object':
        plt.figure(figsize=(8, 6))
        sns.countplot(data=hr_data, x=column, palette='viridis')
        plt.title(f'{column} Count')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.xticks(rotation=45)
        plt.show()
```
